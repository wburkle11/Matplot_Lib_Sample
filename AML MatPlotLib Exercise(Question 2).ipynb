```python
import os 
datapath = '/Users/wburkle11/Desktop/QIS/Applied Machine Learning /Homework 1/archive'
os.path.exists(datapath) 

def load_fire_data(fire_path=datapath):
    csv_path = os.path.join(fire_path, "fire_nrt_V1_96617.csv")
    return pd.read_csv(csv_path)

#Used for loading our fire dataset
```

```python
fire = load_fire_data()
```

```python
#Visualizing the distribution of Brightness temperature I4 as a histogram.
fire.hist('bright_ti4', bins = 50, figsize = (20, 15))
```

```python
#Histograms of all of our column attrubutes in our dataset. 
fire.hist(bins = 50, figsize = (20, 15))
```

```python

fire.plot(kind = "scatter", x="longitude", y="latitude", alpha = 0.2,
          s = 2, figsize = (10, 8), c = "bright_ti4", cmap = plt.get_cmap("jet"), 
          colorbar=True, sharex = False)

fire.plot(kind="scatter", x="longitude", y="latitude", alpha = 0.2, s = 2)

```

```python
#Combining Plots Above
color = fire['bright_ti4']

figure(figsize=(10, 8), dpi=80)
 
plt.style.use('ggplot')
 
plt.title('Saturated Brightness Layed on Long/Lat Datapoints')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.scatter(x=fire['longitude'],y=fire['latitude'], label = 'Unsaturated Brightness', s=5,c= ('b'),alpha=0.6,marker='o', cmap = plt.get_cmap("jet")) #Written with class of Non-Saturated datapoints plotted first
plt.scatter(x=fire['longitude'],y=fire['latitude'], label = 'Saturated Brightness', s=5,c=color,alpha=0.2,marker='h', cmap = plt.get_cmap("jet"))
 
plt.legend() 
plt.tight_layout()
 
plt.show()

print('This plot shows saturated brightness drawn on top of unsaturated brightness datapoints. This plot conveys a signifigant amount of more data compared to the plot generated below. The plot shows the intensity of the fires with respect to its brightness. You can see while there are many more fires around the coastal cities, the fires in the middle of the continent in its desert and forest enviroments the fires are far more intense.')
```
